```python
# Q2(a)Load the data from data.csv file Download data.csv fileOpen this document with 
# ReadSpeaker docReaderinto a dataframe with restaurant ID as the dataframe’s index. 
# Skip the first 50 rows of the file when loading your data (5 points)

import pandas as pd

file = 'data.csv'


df = pd.read_csv(file, index_col='Restaurant ID', skiprows=range(1, 51), encoding='ISO-8859-1')

df
```

```python
# Q2(b) Sort the dataframe first by ‘Votes’ and next by ‘Price range’ in a single step, i.e., 
# entries with equal ‘Votes’ values should be sorted based on their ‘Price range’ values.
# Print the first 15 entries of this sorted dataframe. (5 points)

sort_df = df.sort_values(by=['Votes', 'Price range'], ascending=[False, True])


columns = sort_df[['Votes','Price range']]

print(columns.head(15))
```

```python
# Q2(c)Replace all instances of ‘Indian Rupees(Rs.)’ with ‘INR (Rs.)’ throughout your dataframe. (2 points)


df['Currency'] = df['Currency'].str.replace('Indian Rupees', 'INR (Rs.)')

columns = df['Currency']
print(columns.head(900))
```

```python
# Q2(d)Create a Pandas series that stores the number of restaurants registered with this app in each city.
# Find the city with the highest number of restaurants registered. (5 Points)

number_restaurants = df['City'].value_counts()
highest_restaurants = number_restaurants.idxmax()


print("Number of Restaurants in Each City:")
print(number_restaurants)
print("City with the Highest Number of Restaurants:", highest_restaurants)
```

```python
# Q2(e)Add a new column, “Price Grade” which will have the following values based on the 
# 'Price range' of the restaurant (7 points):

def price_range_to_grade(price_range):
    if price_range == 1:
        return 'Cheap'
    elif price_range == 2:
        return 'Average'
    elif price_range == 3:
        return 'Average'
    elif price_range == 4:
        return 'Expensive'
    else:
        return 'Unknown'

df['Price Grade'] = df['Price range'].apply(price_range_to_grade)

columns = df[['Price range','Price Grade']]

# Print the selected columns
columns
```

```python
# Q2(f)Print the restaurants with rating less than 2. (part 1)

low_rated = df[df['Aggregate rating'] < 2]
columns = low_rated[['Restaurant Name','Aggregate rating']]

print("Restaurants with rating less than 2:")
columns
```

```python
#Q2(f)  Next, drop the entries of those restaurants. (3 Points) (part2)

df = df[df['Aggregate rating'] >= 2]

    
low_rated = df[df['Aggregate rating'] < 2]
columns = low_rated[['Restaurant Name','Aggregate rating']]

columns

```

```python
# Q2(g) Find the number of restaurants where the average cost of two is less than 500 and
# aggregate rating is greater than 3.5. (5 Points)

filter_restaurants = df[(df['Average Cost for two'] < 500) & (df['Aggregate rating'] > 3.5)]

number_restaurants = len(filter_restaurants)

print(f"Number of restaurants with an average cost less than 500 and aggregate rating greater than 3.5 is {number_restaurants} restaurants")


```

```python
# Q2(h)Drop all columns except ‘Restaurant Name’, ‘City’, ‘Price range’, and ‘Votes’ 
# from your dataframe. Add the following entries to this dataframe: (8 points)

new_columns = ['Restaurant Name', 'City', 'Price range', 'Votes']
df = df[new_columns]

new_entries = [
    {'Restaurant Name': 'Amano', 'City': 'New York', 'Price range': 12, 'Votes': 387},
    {'Restaurant Name': 'The Coppola Cafe', 'City': 'New Brunswick', 'Price range': 9, 'Votes': 183}
]

new_entries_df = pd.DataFrame(new_entries)
df = df.append(new_entries_df, ignore_index=True)

print(df)
```

```python
# Q2(i) Convert all column names to UPPERCASE format.
# Save the updated dataframe in an Excel file called ‘output.xlsx’ in a sheet called ‘Restaurants’.
# Ignore writing the index column in this file. Attach this file in your submission. (10 points)


df.columns = df.columns.str.upper()

with pd.ExcelWriter('output.xlsx', engine='openpyxl') as writer:
    df[['RESTAURANT NAME', 'CITY', 'PRICE RANGE', 'VOTES']].to_excel(writer, sheet_name='Restaurants', index=False)


```
